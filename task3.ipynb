```python


#Mention the data point with the help of loops
import matplotlib.pyplot as plt


x = ["aryan","vaibhav","jitesh","kunal"]
y = [40,50,60,70]
plt.plot(x,y,color="g", marker="o")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.scatter(x, y, color="b")
for i in range(len(x)):
    plt.text(i + 0.1, y[i] + 0.1, x[i], fontsize=10)
plt.show()

```

```python


#Add legend in your graph
z = [95,77,86,59]
plt.plot(x,y,color ="r", marker="o",label="mumbai")
plt.plot(x,z,color="b",marker="o",label="hydrabad")
plt.xlabel("name")
plt.ylabel("place")

plt.scatter(x, y, color="y")
plt.legend()
plt.show()
```

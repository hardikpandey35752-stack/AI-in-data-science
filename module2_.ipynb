##merge in Data Frame

```python
#import libraries
import pandas as pd

#create DF's
df1 = pd.DataFrame({"A":[80,85,8], "B":[1,2,3]})
df2 = pd.DataFrame({"A":[80,85,8], "B":[5,6,7]})

#perform merge
pd.merge(df1,df2)
```

##how

```python
#inner
df1 = pd.DataFrame({"A":[1,2,3,4], "B":[11,21,3,44]})
df2 = pd.DataFrame({"A":[1,2,3,5], "B":[10,20,30,45]})
pd.merge(df1,df2,on="A",how="inner")
```

```python
#outer
df1 = pd.DataFrame({"A":[1,2,3,4], "B":[11,21,31,44]})
df2 = pd.DataFrame({"A":[1,2,3,5], "B":[10,20,30,45]})
#perfom outer
pd.merge(df1,df2,on="A",how="outer")
```

```python
#left
df1 = pd.DataFrame({"A":[1,2,3,4], "B":[11,21,31,44]})
df2 = pd.DataFrame({"A":[1,2,3,5], "B":[10,20,30,45]})
#perfom left
pd.merge(df1,df2,on="A",how="left")
```

```python
#right
df1 = pd.DataFrame({"A":[1,2,3,4], "B":[11,21,31,44]})
df2 = pd.DataFrame({"A":[1,2,3,5], "B":[10,20,30,45]})
#perfom right
pd.merge(df1,df2,on="A",how="right")
```

##index

```python
#inner
df1 = pd.DataFrame({"A":[1,2,3,4], "B":[11,21,31,44]})
df2 = pd.DataFrame({"A":[1,2,3,4], "B":[10,20,30,45]})
pd.merge(df1,df2,right_index=True,left_index=True)
```

```python
#inner
df1 = pd.DataFrame({"A":[1,2,3,4], "B":[11,21,31,44]})
df2 = pd.DataFrame({"A":[1,2,3,4], "B":[10,20,30,45]})
pd.merge(df1,df2,right_index=True,left_index=True,suffixes=("_id","_roll no."))
```

```python
sr1 = pd.Series([1,2,3,4])
sr2 = pd.Series([5,6,7,8])
pd.concat([sr1,sr2])
```

```python
#inner
df1 = pd.DataFrame({"A":[1,2,3,4], "B":[11,21,31,44]})
df2 = pd.DataFrame({"A":[1,2,3,5], "B":[10,20,30,45]})

pd.concat([df1,df2])
```

```python
df1 = pd.DataFrame({"A":[1,2,3,4], "B":[11,21,31,44]})
df2 = pd.DataFrame({"A":[1,2,3,5], "B":[10,20,30,45]})

pd.concat([df1,df2],axis=1,keys=["df1","df2"])
```

```python
df1 = pd.DataFrame({"A":[1,2,3,4]})
df2 = pd.DataFrame({"A":[1,2,3,5], "B":[10,20,30,45]})

pd.concat([df1,df2],axis=0,keys=["df1","df2"])
```

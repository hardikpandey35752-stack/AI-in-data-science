```python
#dictionary on matplotlib
import pandas as pd
import matplotlib.pyplot as plt
dict ={
    "marks":[4,5,6,7,8]
}
df=pd.DataFrame(dict)
plt.plot(df["marks"],marker="o",linestyle=":",linewidth=2,color="r")
plt.show()
```

```python

df.index=[7,6,5,4]
display(df)

```

##histogram

```python
plt.hist(df["marks"])
plt.show()
x=["karan","kane","aman","dj","jmin"]
y=[95,87,98,60,99]
plt.grid()
plt.xlabel("name of subject")
```

```python
plt.hist(df["marks"], bins=4, color="y", edgecolor="b", alpha=0.3)
plt.xlabel("marks")
plt.ylabel("frequency")
plt.show()
```

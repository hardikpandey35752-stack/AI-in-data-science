```python
#merge three or more than three dfs
import pandas as pd
df1 = pd.DataFrame({"A":[11,32,43],"B":[44,65,56]})
df2 = pd.DataFrame({"C":[71,82,93],"D":[81,61,91]})
df3 = pd.DataFrame({"E":[14,19,51],"F":[66,67,88]})
pd.merge(df1,df2,df3,on="A", how="inner")
```

```python

#Perform join on df and series
df1 = pd.DataFrame({"A":[11,22,33],"B":[54,75,86]})
df2 = pd.DataFrame({"C":[71,81,92],"D":[16,14,39]})
df1.join(df2)
sr1 = pd.Series([11,22,33],name = "A")
sr2 = pd.Series([54,75,86], name = "B")
sr1.to_frame().join(sr2)
```
